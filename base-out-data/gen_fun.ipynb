```python
series_poly = sp.Poly(f_16.series(x, 0, 20).removeO(), x)
coeffs = series_poly.all_coeffs()                # Ordered from highest power to lowest
numerical_coeffs = [c.evalf() for c in coeffs][::-1]
cvec = np.array(numerical_coeffs)
normalized = cvec / cvec.sum()
normalized
```
